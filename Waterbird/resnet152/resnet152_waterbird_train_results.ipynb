```python
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
```

```python
train_statistics_files = glob("./trained/lr_00001/*.npy")
```

```python
fig = plt.figure(figsize=(20, 70), dpi=200, constrained_layout=True)
fig.suptitle('Resnet152\nLearning rate: 0.00001 | Activation: ReLU', fontsize=20)
subfigs = fig.subfigures(nrows=12, ncols=1)

for statistic_filename, subfig in zip(train_statistics_files, subfigs):
    statistics = np.load(statistic_filename, allow_pickle=True).item()

    sample = statistic_filename.split("\\")[-1][17:-4]

    train_loss = statistics['train_loss']
    train_acc = statistics['train_acc']
    val_loss = statistics['val_loss']
    val_acc = statistics['val_acc']

    subfig.suptitle(f'Augmented samples: {sample}', fontsize=18)
    axs = subfig.subplots(nrows=1, ncols=2)
    epochs = np.arange(1, len(train_acc) + 1)
    axs[0].plot(epochs, train_acc, label='Train accuracy')
    axs[0].plot(epochs, val_acc, label='Val accuracy')
    axs[0].set_xticks(epochs)
    axs[0].set_title('Train and validation accuracy', fontsize=15)
    axs[0].legend()
    axs[0].set_xlabel('Epoch')
    axs[0].set_ylabel('Accuracy')

    axs[1].plot(epochs, train_loss, label='Train loss')
    axs[1].plot(epochs, val_loss, label='Val loss')
    axs[1].set_xticks(epochs)
    axs[1].set_title('Train and validation loss', fontsize=15)
    axs[1].legend()
    axs[1].set_xlabel('Epoch')
    axs[1].set_ylabel('Loss')


plt.show()
```
