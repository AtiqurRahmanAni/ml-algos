```python
import numpy as np
import matplotlib.pyplot as plt
```

$
p(x) = \frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x- \mu)^2}{2\sigma^2}}
$

```python
def gaussian(x: np.ndarray, mu: float, sigma: float):
    y = (1 / np.sqrt(2 * np.pi * sigma * sigma)) * np.exp(-np.square(x - mu) / (2 * sigma * sigma))
    return y


def kl_divergence(y1: np.ndarray, y2: np.ndarray):
    kl = np.sum(y1 * np.log(y1 / y2))
    return kl
```

```python
mu1, sigma1 = 0, 0.6
mu2, sigma2 = 0.1, 0.6
x1 = np.linspace(-2, 2, 1000)
y1 = gaussian(x1, mu1, sigma1)
y2 = gaussian(x1, mu2, sigma2)

kl = kl_divergence(y1, y2)
print(f"KL divergence: {kl:.4f}")

plt.plot(x1, y1, label="Distribution 1")
plt.plot(x1, y2, label="Distribution 2")
plt.legend()
plt.show()
```
